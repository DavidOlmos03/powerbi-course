```python
# service = Service(EdgeChromiumDriverManager().install())
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=chrome_options)
```

```python
driver.get('https://www.olx.in/cars_c84')

# Para el boton
boton = driver.find_element('xpath','//button[@data-aut-id="btnLoadMore"]')

#Espero mientras se carga la información en la página (en segundos)
#Se utiliza el random para evitar comportamientos iguales 
for i in range(3):
    try:
        boton.click()
        sleep(random.uniform(8.0,10.0))
        boton = driver.find_element('xpath','//button[@data-aut-id="btnLoadMore"]')
    except:
        break

```

```python
# Todos los anuncios en la lista
autos = driver.find_elements('xpath','//li[@data-aut-id="itemBox"]')

for auto in autos:
    precio = auto.find_element('xpath','.//span[@data-aut-id="itemPrice"]').text
    print(precio)
    description = auto.find_element('xpath','.//div[@data-aut-id="itemTitle"]').text
    print(description)
```
